```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("h_variation.dat", "r")
xin=[]
yin=[]
y2in=[]
y3in=[]
y4in=[]
y5in=[]


for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
    y2in.append(float(data[2]))
    y3in.append(float(data[3]))
    y4in.append(float(data[4]))
    y5in.append(float(data[5]))
    
filein.close()
```

```python
plt.plot(xin,yin, label=r"N=4")
plt.plot(xin,y2in, label=r"N=8")
plt.plot(xin,y3in, label=r"N=12")
plt.plot(xin,y4in, label=r"N=16")
plt.plot(xin,y5in, label=r"N=20")
plt.xlabel(r'h in a.u.')
plt.ylabel(r'<m> in a.u.')
plt.legend()
plt.savefig('h_variation')
```
