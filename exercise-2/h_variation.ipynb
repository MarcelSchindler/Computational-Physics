```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("h_variation.dat", "r")
xin=[]
yin=[]

for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
filein.close()
```

```python
plt.plot(xin,yin, linewidth=4, label=r"numeric")
plt.xlabel(r'h in a.u.')
plt.ylabel(r'<m> in a.u.')
plt.legend()
plt.savefig('h_variation')
```
