```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("J_variation.dat", "r")
xin=[]
x2in=[]
yin=[]
y2in=[]
y3in=[]
y4in=[]

for line in filein:
    data=line.split()
    x2in.append(float(data[0]))
    xin.append(float(data[1]))
    yin.append(float(data[2]))
    y2in.append(float(data[3]))
    y3in.append(float(data[4]))
    y4in.append(float(data[5]))
filein.close()
```

```python
plt.xlim(1,4)
plt.plot(xin,yin, label=r"numeric")
plt.plot(xin,y2in, label=r"analytic")
plt.xlabel(r'J$^{-1}$ in a.u.')
plt.ylabel(r'<m> in a.u.')
plt.legend()
plt.savefig('J_variation')
```

```python
plt.plot(x2in,y3in, label=r"numeric")
plt.plot(x2in,y4in, label=r"analytic")
plt.xlabel(r'J in a.u.')
plt.ylabel(r'<$\epsilon$> in a.u.')
plt.legend()
plt.savefig('energie_per_site')
```
