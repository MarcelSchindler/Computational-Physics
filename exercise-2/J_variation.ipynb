```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("J_variation.dat", "r")
xin=[]
x2in=[]
yin=[]
y2in=[]
y3in=[]
y4in=[]
y5in=[]
y6in=[]
y7in=[]
y8in=[]
y9in=[]
y10in=[]
y11in=[]
y12in=[]
y13in=[]
y14in=[]
y15in=[]
y16in=[]
y17in=[]
y18in=[]
y19in=[]



for line in filein:
    data=line.split()
    x2in.append(float(data[0]))
    xin.append(float(data[1]))
    yin.append(float(data[2]))
    y2in.append(float(data[3]))
    y3in.append(float(data[4]))
    y4in.append(float(data[5]))
    y5in.append(float(data[6]))
    y6in.append(float(data[7]))
    y7in.append(float(data[8]))
    y8in.append(float(data[9]))
    y9in.append(float(data[10]))
    y10in.append(float(data[11]))
    y11in.append(float(data[12]))
    y12in.append(float(data[13]))
    y13in.append(float(data[14]))
    y14in.append(float(data[15]))
    y15in.append(float(data[16]))
    y16in.append(float(data[17]))
    y17in.append(float(data[18]))
    y18in.append(float(data[19]))
    y19in.append(float(data[20]))
filein.close()
```

```python
plt.xlim(1,4)
plt.plot(xin,yin, label=r"N=4")
plt.plot(xin,y2in, label=r"N=8")
plt.plot(xin,y3in, label=r"N=12")
plt.plot(xin,y4in, label=r"N=16")
plt.plot(xin,y5in, label=r"N=20")

plt.plot(xin,y6in, label=r"analytic")
plt.xlabel(r'J$^{-1}$ in a.u.')
plt.ylabel(r'<|m|> in a.u.')
plt.legend()
plt.savefig('J_variation')
```

```python
plt.xlim(1,4)
plt.plot(xin,y13in, label=r"m")
plt.plot(xin,y5in, label=r"|m|")
plt.xlabel(r'J$^{-1}$ in a.u.')
plt.ylabel(r'<m> in a.u.')
plt.legend()
plt.savefig('J_variation_abs')
```
