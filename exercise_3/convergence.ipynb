```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("convergence.dat", "r")
xin=[]
yin=[]

for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))

    
filein.close()
```

```python
plt.ylim(0,0.1)
#plt.xlim(5,100)
plt.plot(xin,yin, label=r"Leapfrog")
plt.xlabel(r'$N_{md}$')
plt.ylabel(r'$[H(p_0,\phi_0)-H(p,\phi)]/H(p_0,\phi_0)$')
plt.legend()
plt.savefig('convergence')
```
