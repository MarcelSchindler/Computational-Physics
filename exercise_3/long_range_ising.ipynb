```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("long_range_ising.dat", "r")
xin=[]
yin=[]
y1in=[]
y2in=[]
y3in=[]
y4in=[]
y5in=[]
y6in=[]
y7in=[]
y8in=[]
y9in=[]
y10in=[]
y11in=[]
y12in=[]
y13in=[]
y14in=[]
y15in=[]




for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
    y1in.append(float(data[2]))
    y2in.append(float(data[3]))
    y3in.append(float(data[4]))
    y4in.append(float(data[5]))
    y5in.append(float(data[6]))
    y6in.append(float(data[7]))
    y7in.append(float(data[8]))
    y8in.append(float(data[9]))
    y9in.append(float(data[10]))
    y10in.append(float(data[11]))
    y11in.append(float(data[12]))
    y12in.append(float(data[13]))
    y13in.append(float(data[14]))
    y14in.append(float(data[15]))
    y15in.append(float(data[16]))
    
    

    
filein.close()
```

```python

plt.plot(xin,yin, label=r"N=5")
plt.plot(xin,y1in, label=r"N=10")
plt.plot(xin,y2in, label=r"N=15")
plt.plot(xin,y3in, label=r"N=20")


plt.xlabel(r'J')
plt.ylabel(r'$<m>$')
plt.legend()
plt.savefig('magnitization')
```
