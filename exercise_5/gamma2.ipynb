```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("Multigrid_gamma2.dat", "r")
xin=[]
yin=[]
y1in=[]


for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
    y1in.append(float(data[2]))

filein.close()
```

```python
plt.xlim(0,4000)
plt.ylim(-5,20)
plt.scatter(xin,y1in, label=r"$\gamma=2$")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$C(\tau)$')
plt.legend()
plt.savefig('magnitization_square_autocorrelation_gamma2')
```

```python
plt.scatter(xin,yin, label=r"$\gamma=2$")
plt.xlabel(r'i')
plt.ylabel(r'$\phi_i$')
plt.legend()
plt.savefig('magnitization_square_multigrid_gamma2')
```
