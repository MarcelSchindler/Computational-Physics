```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("Metropolis_Hasting.dat", "r")
xin=[]
yin=[]
y1in=[]
y2in=[]


for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
    y1in.append(float(data[2]))
    y2in.append(float(data[3]))

    
filein.close()
```

```python
plt.scatter(xin,yin, label=r"$N_{md}=4$")
plt.xlabel(r'i')
plt.ylabel(r'$<m>$')
plt.legend()
plt.savefig('magnitization')
```

```python
plt.scatter(xin,y1in, label=r"$N_{md}=4$")
plt.xlabel(r'i')
plt.ylabel(r'$<m^2>$')
plt.legend()
plt.savefig('magnitization_square')
```

```python
plt.scatter(xin,y2in, label=r"$N_{md}=4$")
plt.xlabel(r'i')
plt.ylabel(r'$<\epsilon>$')
plt.legend()
plt.savefig('energy')
```
