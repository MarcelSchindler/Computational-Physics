```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("blocking.dat", "r")
xin=[]
y1in=[]
y2in=[]
y3in=[]
y4in=[]
y5in=[]
y6in=[]
y7in=[]
y8in=[]
yin=[]
y9in=[]
y10in=[]
y11in=[]
y12in=[]



for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    y1in.append(float(data[1])) 
    y2in.append(float(data[2]))
    y3in.append(float(data[3]))
    y4in.append(float(data[4]))
    y5in.append(float(data[5]))
    y6in.append(float(data[6]))
    y7in.append(float(data[7]))
    y8in.append(float(data[8]))
    y9in.append(float(data[9]))
    y10in.append(float(data[10]))
    y11in.append(float(data[11]))
    y12in.append(float(data[12]))
    
filein.close()
```

```python
plt.xlim(0,50)
plt.ylim(-0.01,0.01)
plt.scatter(xin,y1in, label=r"b=2")
plt.scatter(xin,y3in, label=r"b=4")
plt.scatter(xin,y5in, label=r"b=8")
plt.scatter(xin,y7in, label=r"b=16")
plt.scatter(xin,y9in, label=r"b=32")
plt.scatter(xin,y11in, label=r"b=64")
plt.xlabel(r'$\tau$')
plt.ylabel(r'$C(\tau)$')
plt.legend()
plt.savefig('blocking_autokorrelation')
```

```python
plt.xlim(0,120)
plt.ylim(0.2,0.4)
plt.scatter(xin,y2in, label=r"b=2")
plt.scatter(xin,y4in, label=r"b=4")
plt.scatter(xin,y6in, label=r"b=8")
plt.scatter(xin,y8in, label=r"b=16")
plt.scatter(xin,y10in, label=r"b=32")
plt.scatter(xin,y12in, label=r"b=64")
plt.xlabel(r'$i$')
plt.ylabel(r'$\sigma_i$')
plt.legend()
plt.savefig('blocking_standard_deviation')
```
