```python
import pylab as plt
import sys
import pandas as pd
import numpy as np
import math as m

print(sys.version)
```

```python
filein=open("comparison.dat", "r")
xin=[]
yin=[]
y1in=[]
y2in=[]
y3in=[]

for line in filein:
    data=line.split()
    xin.append(float(data[0]))
    yin.append(float(data[1]))
    y1in.append(float(data[2]))
    y2in.append(float(data[3]))
    y3in.append(float(data[4]))

    
filein.close()
```

```python
plt.scatter(xin,yin, label=r"$N_{md}=4$")
plt.xlabel(r'i')
plt.ylabel(r'$\phi_i$')
plt.legend()
plt.savefig('comparison_MC')
```

```python
plt.scatter(xin,y1in, label=r"$N_{md}=100$")
plt.xlabel(r'i')
plt.ylabel(r'$\phi_i$')
plt.legend()
plt.savefig('comparison_MC')
```

```python
plt.scatter(xin,y3in, label=r"$N_{md}=100$")
plt.xlabel(r'i')
plt.ylabel(r'$<m(\phi_i)>$')
plt.legend()
plt.savefig('comparison_magnitization_100')
```

```python
plt.scatter(xin,y2in, label=r"$N_{md}=4$")
plt.xlabel(r'i')
plt.ylabel(r'$<m(\phi_i)>$')
plt.legend()
plt.savefig('comparison_magnitization_4')
```
